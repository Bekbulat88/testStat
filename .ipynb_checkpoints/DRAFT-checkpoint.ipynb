8. Во вкладке "Данные АБ-тестов" результаты трех несвязанных АБ тестов для ARPU (общая выручка/общее количество пользователей).
Посмотрите на результаты тестов и интерпретируйте их. Напишите значения p-value, которые вы получили.
Подготовьте выводы и рекомендации. 

experiment_num - номер эксперимента
experiment_group - группа, в которую попал пользователь
user_id - id пользователя
revenue - выручка, которую сгенерировал пользователь, купив платную услугу продв

получается была какая то фича на тестовую группу и запущен аб тест, по итогу есть эти показатели. мне надо расчитать ARPUtest, ARPUcontrol по 3 раза.
затемижения


```python
dfAB = pd.read_csv('Данные для тестового задания - Данные АБ тестов.csv')
```

```python
dfAB.head()
```

```python
H0: arpu same
H1: ARPUtest not same ARPUcontrol
```

```python
import statsmodels.api as sm
from scipy.stats import norm
from scipy.stats import mannwhitneyu
from scipy.stats import shapiro
```

```python
group_A = dfAB[dfAB['experiment_group'] == 'control']['revenue']
group_B = dfAB[dfAB['experiment_group'] == 'test']['revenue']
```

```python

norm_control_p = shapiro(group_A).pvalue
norm_test_p = shapiro(group_B).pvalue
```

```python
mannwhitney_results = []
for experiment in dfAB['experiment_num'].unique(): # [1,2,3]
    exp_data = dfAB[dfAB['experiment_num'] == experiment]
    control = exp_data[exp_data['experiment_group'] == 'control']['revenue']
    test = exp_data[exp_data['experiment_group'] == 'test']['revenue']
    
    # Mann-Whitney U test
    u_stat, p_value = mannwhitneyu(control, test, alternative='two-sided')
    
    # Mean revenue (for reference)
    ARPU_control = control.sum() / control.nunique()
    ARPU_test = test.sum() / test.nunique()
    
    # Append results
    mannwhitney_results.append({
        'experiment_num': experiment,
        'ARPU_control': ARPU_control,
        'ARPU_test': ARPU_test,
        'p_value': p_value
    })

```

```python
mannwhitney_results_df = pd.DataFrame(mannwhitney_results)
mannwhitney_results_df
```
